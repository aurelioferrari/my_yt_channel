```python
df_total = pd.concat([df_gow, df_re_2, df_mk])
df_total
```

```python
top10_views = df_total.groupby(['Título do vídeo'])['Visualizações'].sum().sort_values(ascending=True).tail(10)
top10_imps = df_total.groupby(['Título do vídeo'])['Impressões'].sum().sort_values(ascending=True).tail(10)
```

```python
plt.style.use('dark_background')
top10_views.plot.barh(color='green')
plt.xlabel('Views', size=20, color='white')
plt.ylabel('Video Title', size=20, color='white')

plt.show()
```

```python
plt.style.use('dark_background')
top10_imps.plot.barh()
plt.xlabel('Impressions', size=20, color='white')
plt.ylabel('Video Title', size=20, color='white')
plt.show()
```

```python

top10_click = df_total.groupby(['Título do vídeo'])['Taxa de cliques de impressões (%)'].sum().sort_values(ascending=True).tail(10)
top10_click
```

```python
plt.style.use('dark_background')
top10_click.plot.barh(color='darkred')
plt.xlabel('clicks %', size=20, color='white')
plt.ylabel('Video Title', size=20, color='white')
plt.show()
```
